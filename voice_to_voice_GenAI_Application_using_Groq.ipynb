```python
import gradio as gr       # create U/I in huggingfaces
import os                 # to call key
import whisper            # Transcribe the Audio
from groq import Groq     # provide API
from gtts import gTTS     # convert Text to Voice

# Initialize Whisper model for transcription
model = whisper.load_model("base")

# Set up Groq API
client = Groq(api_key=os.environ.get("GROQ_API_Key"))


# Function to query the LLM using Groq API
def get_llm_response(input_text):
    chat_completion = client.chat.completions.create(
        messages=[{
            "role": "user",
            "content": input_text,
        }],
        model="llama3-8b-8192",
    )
    return chat_completion.choices[0].message.content


# Function to convert text to speech using gTTS
def text_to_speech(text,output_audio="output_audio.mp3"):
    tts = gTTS(text)
    tts.save(output_audio)
    return output_audio


def chatbot(audio):
  result=model.transcribe(audio)
  user_text=result['text']
  response_text=get_llm_response(user_text)
  output_audio=text_to_speech(response_text)
  return response_text,output_audio


# Create Gradio interface for microphone input
#iface = gr.Interface(
    #fn=chatbot,
    #inputs=gr.Audio(type="filepath"),  # Capturing audio from the microphone
    #outputs=[gr.Textbox(),gr.Audio(type="filepath")],  # Outputting audio file path
    #live=True
#)

#iface.launch()

# Create Gradio interface

iface = gr.Interface(
    fn=chatbot,
    inputs=gr.Audio(type="filepath", label="ğŸ¤ Microphone Input"),  # Capture audio from the microphone
    outputs=[
        gr.Textbox(label="ğŸ“ Transcribed Text"),  # Display transcribed text
        gr.Audio(type="filepath", label="ğŸ”Š Audio Response")  # Output the generated audio response
    ],
    title="ğŸ™ï¸ Voice-Based Chatbot - by Shoaib Aqeel",
    description="This chatbot accepts voice input, transcribes it to text, processes the response, "
                "and provides an audio output. Speak into the microphone to interact with the chatbot.",
    live=True
)

# Launch the interface
iface.launch()
```
